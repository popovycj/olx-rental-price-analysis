```python
import pandas as pd
```

```python
column_names = ['title', 'price', 'price_currency', 'description', 'views_count', 'created_at', 'options', 'location', 'region', 
                'seller_name', 'seller_registration_date', 'seller_link', 'announcement_link']

df = pd.read_csv('data.csv', names=column_names, header=None)

df.tail()
```

```python
# Check uniqueness

df.drop_duplicates(subset=['announcement_link'], inplace=True)

len(df) == df['announcement_link'].nunique()
```

```python
# Remove views_count column because of its unrepresentativeness and uselessness

df.drop('views_count', axis=1, inplace=True)

df.columns
```

```python
# Convert any other currencies to UAH

USD_to_UAH = 37.5
EUR_to_UAH = 40.5

df.loc[df['price_currency'] == 'USD', 'price'] = df[df['price_currency'] == 'USD']['price'].apply(lambda price: float(price) * USD_to_UAH)
df.loc[df['price_currency'] == 'EUR', 'price'] = df[df['price_currency'] == 'EUR']['price'].apply(lambda price: float(price) * EUR_to_UAH)

df[df['price_currency'] == 'EUR'].head()
```

```python
# Remove price_currency column

df.drop('price_currency', axis=1, inplace=True)

df.head()
```

```python
# Remove blank announcements

df = df[df['price'] != 'None']

df.tail()
```

```python
from random import randint
import ast

def parse_options(str_options):
    single_keys = ['Тип', 'Додаткова інформація', 'Тип будинку',
                   'Поверх', 'Поверховість', 'Загальна площа', 'Площа кухні',
                   'Тип стін', 'Кількість кімнат', 'Планування', 'Cанвузол',
                   'Опалення', 'Ремонт', 'Меблювання']

    multiple_keys = ['Додаткова інформація', 'Домашні улюбленці', 'Побутова техніка',
                     'Мультимедіа', 'Комфорт', 'Комунікації', 'Інфраструктура (до 500 метрів)',
                     'Ландшафт (до 1 км.)', 'Додатково']

    options = {key: [] if key in multiple_keys else None for key in single_keys + multiple_keys}

    try:
        array_of_options = ast.literal_eval(str_options)
        if not isinstance(array_of_options, list):
            return options

        options['Тип'] = array_of_options.pop(0)

        while array_of_options and array_of_options[0] in ['Без комісії', 'Готовий співпрацювати з ріелторами', 'Для спільної оренди']:
            options['Додаткова інформація'].append(array_of_options.pop(0))

        for option in array_of_options:
            key, value = option.split(': ', 1)
            if key in ['Побутова техніка', 'Мультимедіа', 'Комфорт', 'Комунікації', 'Інфраструктура (до 500 метрів)', 'Ландшафт (до 1 км.)', 'Додатково']:
                value = value.split(', ')
            elif key == 'Домашні улюбленці':
                value = [x.strip(', ')[:-2] for x in value.split('Так, ') if len(x) > 1] if 'Ні' not in value else []
            else:
                value = value.replace('м²', '').replace('кімната', '').replace('кімнати', '').replace('кімнат', '').strip()
            options[key] = value
    except ValueError:
        pass

    return options


index = randint(1, 8000)
# index = 3352
print(f'INDEX: {index}')
options = df['options'][index]
print('-'*20 + 'NOT CLEARER' + '-'*20)
print(options)
print('-'*20 + 'CLEARER' + '-'*20)
print(parse_options(options))
```

```python
df_options = pd.concat([df['options'].apply(parse_options)], axis=1)
df_options
```

```python
# Extract all options keys as new columns

df['options'] = df['options'].apply(parse_options)

for key in df['options'][0].keys():
    df[key] = df['options'].apply(lambda x: x[key])

df = df.drop('options', axis=1)

df.columns
```

```python
# Convert all prices to float, remove invalid prices

df['price'] = pd.to_numeric(df['price'], errors='coerce')
df.dropna(subset=['price'], inplace=True)

df['price'].apply(lambda x: '{:.2f}'.format(x)).unique()
```

```python
df = df[df['price'] <= 1_000_000]

df.nlargest(15, 'price')
```

```python
df = df[df['price'] >= 1_500]

df.nsmallest(15, 'price')
```
